```python
#Metodo index()

#Un string s una cadena de caracteres por ejemplo hola = h = 0 | o = 1 | l = 2 | a = 3 |

#1
#Para buscar el indice se tiene que insertar la variable y luego entre llaves el numero de indice deseado
mi_texto = "Esta es una prueba"
resultado = mi_texto[0]
print(f"1 = {resultado}")

#2
#Se puede explorar dependiendo la cantidad de caracteres que tenga, incluidos los espacios
mi_texto = "Esta es una prueba"
resultado = mi_texto[9]
print(f"2 = {resultado}")

#De aatras para adelante
mi_texto = "Esta es una prueba"
resultado = mi_texto[-4]
print(f"3 = {resultado}")

#De esta forma, si deseamos buscar la "n", por ejemplo, nos indicara que numero de indice tiene 
mi_texto = "Esta es una prueba"
resultado = mi_texto.index("n")
print(f"4 = {resultado}")

#Podemos buscar por palabras, te dira desde donde comienza el substring
mi_texto = "Esta es una prueba"
resultado = mi_texto.index("prueba")
print(f"5 comienza desde = {resultado}")

#Siempre va a buscar la primera a que aparezca, pero tambien le podemos indicar desde que indice empezar
mi_texto = "Esta es una prueba"
resultado = mi_texto.index("a",5)
print(f"6 = {resultado}")

#Tambien le podes indicar desde donde hasta donde
mi_texto = "Esta es una prueba"
resultado = mi_texto.index("a",5,12)
print(f"7 = {resultado}")

#Reverse index, cuenta de derecha a izquierda
mi_texto = "Esta es una prueba"
resultado = mi_texto.rindex("a")
print(f"8 = {resultado}")
```

```python
#Slicing

#Manera de extraer caracteres desde el indice hasta : otro indice

texto = "ABCDEFGHIJKLM"
fragmento = texto[2]
print(f"1. {fragmento}")

#Segundo factor, usando :. No incluye el ultimo caracter
texto = "ABCDEFGHIJKLM"
fragmento = texto[2:5]
print(f"2. {fragmento}")

#Si se deja vacio interpreta que desde el indice q le indicaste hasta el final
texto = "ABCDEFGHIJKLM"
fragmento = texto[2:]
print(f"3. {fragmento}")

#Lo mismo al reves
texto = "ABCDEFGHIJKLM"
fragmento = texto[:5]
print(f"4. {fragmento}")

#tercer factor, cada cuantos caracteres ir extrayendo
texto = "ABCDEFGHIJKLM"
fragmento = texto[2:10:2]
print(f"5. {fragmento}")

#Tambien se puede dejar libre, pero que vaya salteando por ejemplo de 3 en 3
texto = "ABCDEFGHIJKLM"
fragmento = texto[::3]
print(f"6. {fragmento}")

#Tambien se puede hacer en modo inverso con un numero negativo
texto = "ABCDEFGHIJKLM"
fragmento = texto[::-1]
print(f"7. {fragmento}")
```

```python
#upper() y lower() y split() y join() y find() y replace()


#Upper
texto = "Este es el texto de Luciano"
resultado = texto.upper()
print(f"1. {resultado}")

#podemos determinar un indice en particular
texto = "Este es el texto de Luciano"
resultado = texto[2].upper()
print(f"2. {resultado}")

#lower
texto = "Este es el texto de Luciano"
resultado = texto.lower()
print(f"3. {resultado}")

#split
texto = "Este es el texto de Luciano"
resultado = texto.split()
print(f"4. lo separa en forma de lista {resultado}")

#split utilizando un caracter en forma de separacion
texto = "Este es el texto de Luciano"
resultado = texto.split("t")
print(f"5. {resultado}")



#Metodo Join
a = "Aprender"
b = "Pyhthon"
c = "es"
d = "genial"
#Se ingresan las variables, se utiliza la variable e como espacio, y con la sentencia join() se unen con el caracter " " 
e = " ".join([a,b,c,d])
print(f"6. {e}") 


#metodo find, 
texto = "Este es el texto de Luciano"
resultado = texto.find("texto")
print(f"7. {resultado}")

#Replace para reemplazar por ejemplo palabras o caracteres
texto = "Este es el texto de Luciano"
resultado = texto.replace("Luciano", "Todos")
print(f"8. {resultado}")





```

```python
#Diccionarios
#Los diccionarios se escriben entre {}
#Es como un diccionario, que tiene una coleccion de conceptos y al lado su definicion
#en python es una coleccion de claves y a su vez tiene su valor.

diccionario = {'c1':'valor1','c2':'valor2'}
print(type(diccionario))
print(diccionario)

#busqueda dentro de los diccionarios
resultado = diccionario['c1'] #que tengo alojado en c1
print(resultado)

#Creamos un diccionario con el nombre appellido peso y talla
cliente ={'nombre':'Juan','apellido':'Fuentes','peso':88,'talla':1.73}
consulta = (cliente['talla']) #por ejemplo, le preguntamos: ¿Que tengo alojado dentro del diccionario en "talla"
print(consulta)

dic = {'c1':55,'c2':[10,20,30],'c3':{'s1':100,'s2':200}} #Tenemos un diccionario, con un valor, una lista y otro diccionario
print(dic['c2']) #aqui consultamos que valor hay en el concepto 2, que en este caso es una lista
print(dic['c2'][1]) #y dentro de esta lista que hay en el indice 1
print(dic['c3']) #lo mismo dentro de los diccionarios
print(dic['c3']['s2']) #que hay en el valor

dic = {'c1':['a','b','c'],'c2':['d','e','f']}
resultado = (dic['c2'][1].upper())
print(resultado)

dic = {1:'a',2:'b'}
print(dic)

#Como agregar otro concepto y su valor dentro del diccionario
dic[3] = 'c'
print(dic)


#como sobreescribir dentro del diccionario
dic[2] = 'B'
print(dic)


#como concer todas las claves
print(dic.keys())

#solo los valores
print(dic.values())

#todo lo que hay dentro de un diccionario
print(dic.items())
```

```python
#Tuplas
#Son inmutables, se escribe igual que las listas, en vez de corchetes se usa con parentesis
#por que tuplas?
#ocupan menos espacio
#cuando queremos crear estructuras que no sean modificadas.


mi_tuple = (1,2,3,4)
print(type(mi_tuple))

t =(5,5.6,"ff") # se pueden alojar cualquier tipo de objetos
print(type(t))

#posiciones
print(mi_tuple[-2])

#los tuples no son modficables

#pero si se puede anidar

#tuple dentro de una tuple y ver posiciones
mi_tuple = (1,2,(10,20),4)
print(mi_tuple[2][0])

#se puede convertir en lista
mi_tuple = list(mi_tuple)
print(type(mi_tuple))

#y se puede volver a convertir en tupla
mi_tuple = tuple(mi_tuple)
print(type(mi_tuple))



#Asignar el contenido de un tuple a diferentes variables
t = (1,2,3)
x,y,z = t
print(x,y,z)#siempre y cuando tengan la cantidad exacta de valores

t = (1,2,3,1)
print(len(t)) #cuantos

print(t.count(1)) #ccantidad de apariciones del valor

print(t.index(2)) #numero de indice que se encuentra ese valor
```

```python
#Set es una coleccion de elementos
#se puede usar utilizando set([1,3,5]) o solo entre {}
#Solo admiten elementos unicos, es la razon por la cual se usa.
#No estan ordenados en indice
#Sus elementos son inmutables, no se pueden incluir listas tampoco diccionarios

mi_set = set([1,2,3,4,5])
print(type(mi_set))
print(mi_set)

otro_set = {1,2,3,4,5}
print(type(otro_set))
print(otro_set)


mi_set = set([1,2,3,4,3,3,3,35]) #podes agregar repeticiones pero jamas lo va a tomar porq no puede haber elementos repetidos
print(type(mi_set))
print(mi_set)

mi_set = set([1,2,(3,4,5),4,5]) #Si se puede agregar tuples, al igual que los sets tambien las tuples son inmutables.
print(type(mi_set))
print(mi_set)

mi_set = set([1,2,3,4,5])
print(type(mi_set))
print(len(mi_set)) #fucncion len


mi_set = set([1,2,3,4,5])
print(type(mi_set))
print(2 in mi_set) #2 esta en mi_set?

#Union de sets
s1 = {1,2,3} #set 1
s2 = {3,4,5} #set 2
s3 = s1.union(s2) # union de set 1 con set 2
print(s3) #y como tanto como en el s1, como en el s2 hay elementos repetidos, va a contar solamente 5 elementos, porque no se pueden repetir

#metodo .add
s1 = {1,2,3} #set 1
s1.add(4)
print(s1)

#metodo .remove
s1 = {1,2,3} #set 1
s1.remove(2)
print(s1)

#metodo .discard
s1 = {1,2,3}
s1.discard(6)
print(s1)

#metodo pop() va a eliminar un elemento aleatorio porque no hay indice en el set
s1 = {1,2,3}
sorteo = s1.pop() 
print(sorteo) # sirve por ej para hacer sorteos


#metodo .clear
s1 = {1,2,3}
s1.clear()
print(s1)




```

```python
#Booleanos
#solo son 2 valores, true or false
#Se pueden hacer comparaciones y el resultado de esa comparativa va a ser true or false
#Son la base de la inteligencia artificial, son logicas.

var1 = True
var2 = False

print(type(var1))
print(var1)

numero = 5 > 2+3 #comparacion y va a dar falso porq es igual
print(type(numero))
print(numero)

numero = 5 == 2+3 # == es igual 
print(type(numero))
print(numero)

numero = 5 >= 2+3 #es mayor o igual
print(type(numero))
print(numero)

numero = 5 != 2+3 #Son diferentes? no
print(type(numero))
print(numero)

numero = bool(5>6) # es lo mismo que lo anterior
print(numero)

#Se encuentra el 5 en esta lista?
lista = [1,2,3,4]
control = 5 in lista
print(type(control))
print(control)

#No hay 5 en esta lista?
lista = [1,2,3,4]
control = 5 not in lista
print(type(control))
print(control)
```

```python
texto = input("Ingresa un texto a eleccion: ")
letras = [] #donde se van a alojar las letras

texto = texto.lower()

letras.append(input("Ingresa la primera letra: ").lower())
letras.append(input("Ingresa la segunda letra: ").lower())
letras.append(input("Ingresa la tercera letra: ").lower())

print("\n")
print("CANTIDAD DE LETRAS")
cantidad_letras1 = texto.count(letras[0])
cantidad_letras2 = texto.count(letras[1])
cantidad_letras3 = texto.count(letras[2])

print(f"Hemos encontrado la letra {letras[0]} repetida {cantidad_letras1} veces")
print(f"Hemos encontrado la letra {letras[1]} repetida {cantidad_letras2} veces")
print(f"Hemos encontrado la letra {letras[2]} repetida {cantidad_letras3} veces")

print("\n")
print("CANTIDAD DE PALABRAS")

palabras = texto.split()
print(f"Hemos encontrado {len(palabras)} palabras en tu texto")


print("\n")
print("LETRAS DE INICIO Y DE FIN")
letra_inicio = texto[0]
letra_final = texto[-1]
print(f"La letra inicial es '{letra_inicio}' y la letra final es '{letra_final}'")


print("\n")
print("TEXTO INVERTIDO")
palabras.reverse()
texto_invertido = " ".join(palabras)
print(f"Si ordenamos tu texto al reves va a decir: '{texto_invertido}'")

print("\n")
print("BUSCANDO LA PALABRA PYTHON")
buscar_python = "python" in texto
dic = {True:"si",False:"no"}
print(f"La palabra 'python' {dic[buscar_python]} se encuentra en el texto")





```
